```python
np.arange(1, 13)
```

```python
# green = 1-9, orange = 10-12
crate = np.arange(1, 13)

N = 10
res = []
count = 0
for i in range(N):
    pick = np.random.choice(crate)
    res.append(pick)
    if pick < 10:
        count += 1
    
print(res)
print('green count:', count)
```

Increase N

```python
Ns = [10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000, 500000]

green_chance = []
for N in Ns:
    count = 0
    for i in range(N):
        pick = np.random.choice(crate)
        if pick < 10:
            count += 1
    green_chance.append(count/N)
    
plt.plot(Ns, green_chance)
plt.plot([min(Ns), max(Ns)], [0.75, 0.75], 'r--')
```

# Monty Hall Problem

```python
# Number of repetitions
n = 1000

doors = [1, 2, 3]

pick = np.random.choice(doors, 1000)
plt.hist(pick)
```

```python
results = np.zeros((0, 3))

for i in range(n):
    
    # Set reward to a door
    reward = np.random.choice(doors, 1)[0]
    
    # Contestant's chosen door
    chosen = np.random.choice(doors, 1)[0]
    
    # The door a host opened
    hostdoors = [1, 2, 3]
    hostdoors.remove(reward) # host cannot open the door w/ reward
    if chosen != reward:
        hostdoors.remove(chosen) # host cannot open the chosen door
    host = np.random.choice(hostdoors, 1)[0]
    
    results = np.vstack((results, [reward, chosen, host]))
```

```python
import matplotlib

font = {'family' : 'consolas',
#         'weight' : 'bold',
        'size'   : 20}

matplotlib.rc('font', **font)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3)
fig.tight_layout(rect=[0.0, 0.0, 4, 2])

n1 = 10
plt.subplot(1,3,1)
plt.plot(results[:n1,0], 'ro')
plt.plot(results[:n1,1], 'bx')
# plt.plot(results[:n1,2], 'g+')
plt.title('First 10 results: o=reward; x=chosen')

ymin, ymax = 0, 1

# Calculate cumulative ratio of winning by staying
stay_wins = (results[:,0] == results[:,1])*1
cum_wins = np.cumsum(stay_wins)
cum_ratio = cum_wins / np.arange(1, (n+1))

plt.subplot(1,3,2)
plt.plot(cum_ratio)
plt.title('Chance: win if stay')
plt.ylim((ymin, ymax))

# Calculate cumulative ratio of winning by changing
change_wins = (results[:,0] != results[:,1])*(results[:,0] != results[:,2]) # reward != chosen & reward != host # the later is not necessary, but anyway.
cum_cwins = np.cumsum(change_wins)
cum_cratio = cum_cwins / np.arange(1, (n+1))

plt.subplot(1,3,3)
plt.plot(cum_cratio)
plt.title('Chance: win if change')
plt.ylim((ymin, ymax))
```

# Variance vs Entropy

Distribution 1

```python
def pmf1(x):
        
    p = 0
    if x == 0:
        p = 1/4
    elif x == 1:
        p = 1/4
    elif x == 2:
        p = 1/4
    elif x == 3:
        p = 1/4
        
    return p
```

```python
xs = np.linspace(-1, 10, 100)
plt.plot(xs, [pmf1(i) for i in xs], 'o')
```

```python
# Distribution 1
def cdf1(x):
    return (x >=0) * 1/4 + (x >= 1) * 1/4 + (x >=2) * 1/4 + (x >=3) * 1/4
```

```python
xs = np.linspace(-1, 10, 100)
plt.plot(xs, cdf1(xs))
plt.title('Distribution 1')
plt.xlabel('x')
plt.ylabel('CDF')
```

```python
np.sum((xs - EX1)**2 * pxs)
```
