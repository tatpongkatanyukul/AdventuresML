# Infinity Effect

```python
xs =np.linspace(0, 800)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)
fig.tight_layout(rect=[0.1, 0.1, 3, 3], h_pad=2, w_pad=2)

ymin, ymax = 0, 800
plt.subplot(2,1,1)
plt.plot(xs, xs, 'r', linewidth=2)
plt.ylim((ymin, ymax))

plt.subplot(2,1,2)
plt.plot(xs, np.log(np.exp(xs)), 'b', linewidth=2)
plt.ylim((ymin, ymax))
```

```python
np.log(np.exp(800))
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
fig.tight_layout(rect=[0.1, 0.1, 3, 3], h_pad=2, w_pad=2)

plt.subplot(1,2,1)
plt.plot(xs, xs + 100, 'r', linewidth=2)
plt.plot(xs, np.log(np.exp(xs)), 'b', linewidth=2)

plt.subplot(1,2,2)
plt.plot(xs, xs, 'r', linewidth=2)
plt.plot(xs, np.log(np.exp(xs)) + 100, 'b', linewidth=2)

```

```python
xs = np.linspace(100, 800)
plt.plot(xs, xs/xs, 'r', linewidth=4)
plt.plot(xs, xs/np.log(np.exp(xs)), 'b', linewidth=1.5)
```

# NaN

```python
def softmax(v):
    ev = np.exp(v)
    return ev/np.sum(ev)
```

```python
x = np.array([1, 2, 5])

softmax(x)
```

```python
x = np.array([1, -8, 0])

softmax(x)
```

```python
x = np.array([-11, -8, 0])

softmax(x)
```

```python
x = np.array([-11, -8, -3])

softmax(x)
```

```python
softmax(np.array([1000, 2000, 5000]))
```

# Underflow

```python
N = 20000

p = 1/N

word_p = 1

for i in range(100):
    word_p *= p

print(word_p)
```
