```python
sigma = 1

Ns = np.array([10, 50, 100, 500, 1000])

t1 = time.time()
size = 5000
varofmu = []
for n in Ns:
    x = np.random.normal(0, sigma, n*size).reshape((n, -1))
    print(x.shape)
    varofmu.append(np.var(np.mean(x, axis=0)))
    
t2 = time.time()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
fig.tight_layout(rect=[0.1, 0.1, 1, 1], h_pad=3, w_pad=2)

plt.plot(Ns, np.sqrt(varofmu), 'b-', label='stdev(mu(x))')
plt.plot(Ns, sigma/np.sqrt(Ns), 'r--', label='sigma/sqrt(N)')
plt.legend()
plt.title('sigma=%0.1f'%sigma)
plt.xlabel('N')
plt.ylabel('standard error of mean')
```

# Combo plot

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
fig.tight_layout(rect=[0.1, 0.1, 3, 3], h_pad=6, w_pad=2)

Ns = np.array([10, 50, 100, 500, 1000])
i = 1

for size in [10, 50, 250]:
    for sigma in [0.2, 1, 5]:
        t1 = time.time()
        varofmu = []
        for n in Ns:
            x = np.random.normal(0, sigma, n*size).reshape((n, -1))
#             print(x.shape)
            varofmu.append(np.var(np.mean(x, axis=0)))

        t2 = time.time()

        plt.subplot(3,3, i)
        plt.plot(Ns, np.sqrt(varofmu), 'b-', label='stdev(mu(x))')
        plt.plot(Ns, sigma/np.sqrt(Ns), 'r--', label='sigma/sqrt(N)')
        plt.legend()
        plt.title('sigma=%0.1f (%d repeats)'%(sigma, size))
        plt.xlabel('N')
        plt.ylabel('standard error of mean')
        
        i += 1
```
